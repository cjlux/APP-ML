## Example avec le data set `beans`

voir https://www.tensorflow.org/datasets/catalog/beans

```python
import tensorflow_datasets as tfds
import numpy as np
```

```python
data = tfds.load('beans')
print("clefs trouvées pour les données:", data.keys())
```

on peut utiliser les clefs `train` et `validation` pour charger les données dans les tableaux `numpy`:

```python
(img_train, label_train), (img_test, label_test) = tfds.as_numpy(tfds.load(
    'beans',
    split=['train', 'validation'],
    batch_size=-1,
    as_supervised=True))
```

```python
img_train.shape, label_train.shape
```

=> 1034 images d'entraînement 500 x 500 pixels, en couleur (RGB) et 
1034 labels

```python
img_test.shape, label_test.shape
```

`set` permet de créer l'ensemble des labels (occurence unique de chaque valeur) :

```python
labels = set(label_train)
labels
```
